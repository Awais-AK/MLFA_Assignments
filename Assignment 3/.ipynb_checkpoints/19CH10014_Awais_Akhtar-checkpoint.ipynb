```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('traindata.csv')
m=len(data.iloc[:,1])
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
import string
punc = str.maketrans('','',string.punctuation)

sentences = []

for line in data.iloc[:,1]:
    dist_words = [ word.translate(punc).lower() for word in line.split() ]
    sentences.append(dist_words)    
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
mod_sents = []
filtered_data = []
for line in sentences:
    mod_data = []
    for word in line:
        if word not in stop_words:
            mod_data.append(ps.stem(word))
            filtered_data.append(word)
    mod_sents.append(mod_data)
print("Entire dataset after removing punctuations\n")    
print(filtered_data)
print("\nAfter stemming\n")
print(mod_sents)
```

```python
#Finding vocabulary using sklearn. 

from sklearn.feature_extraction.text import CountVectorizer
# list of text documents

# create the transform
vectorizer = CountVectorizer()

# tokenize and build vocab
vectorizer.fit(filtered_data)

# summarize
print(vectorizer.vocabulary_)

# encode document

vector = vectorizer.transform(stop_words)

# summarize encoded vector
print(vector.shape)
print(type(vector))
print(vector.toarray())
```

```python
covid = {}
science = {}
business = {}
sports = {}

n_covid = 0
n_science = 0
n_business = 0
n_sports = 0

for i in range(m):
    if data.iloc[i,0] == 'covid':
        n_covid += 1
        words = mod_sents[i]
        for single in words:
            if single not in covid.keys():
                covid.update({single:1})
            else:
                covid[single]+=1
    elif data.iloc[i,0] == 'science':
        n_science += 1
        words = mod_sents[i]
        for single in words:
            if single not in science.keys():
                science.update({single:1})
            else:
                science[single]+=1
    elif data.iloc[i,0] == 'business':
        n_business += 1
        words = mod_sents[i]
        for single in words:
            if single not in business.keys():
                business.update({single:1})
            else:
                business[single]+=1
    elif data.iloc[i,0] == 'sports':
        n_sports += 1
        words = mod_sents[i]
        for single in words:
            if single not in sports.keys():
                sports.update({single:1})
            else:
                sports[single]+=1
print("No. of occurence of each class is \nCovid : {}\nScience : {}\nBusiness : {}\nSports : {} ".format(n_covid,n_science,n_business,n_sports))                
```

```python
#Calculating the prior probability
covid_pr = n_covid/m
science_pr = n_science/m
sports_pr = n_sports/m
business_pr = n_business/m
print("Prior probablity of each class is \nCovid : {}\nScience : {}\nBusiness : {}\nSports : {} ".format(covid_pr,science_pr,business_pr,sports_pr))
```

```python
#using our vocabulary in test set
test_set = pd.read_csv('testdata.csv')
test_sample = np.array(test_set.iloc[:,0:2])
test_size = len(test_sample)
class_condition = []

expected_label = []

labels = {0:'covid', 1:'sports', 2:'business', 3:'science'}
t_sentences = []
for line in test_sample[:,1]:
    dist_words = [ word.translate(punc).lower() for word in line.split() ]
    t_sentences.append(dist_words)
t_sentences 
```

```python
t_mod = []
for line in t_sentences:
    mod_data = []
    for word in line:
        if word not in stop_words:
            mod_data.append(ps.stem(word))
    t_mod.append(mod_data)
t_mod
```

```python
for i in range(test_size):
    covid_prob = 1
    sports_prob = 1
    science_prob = 1
    business_prob = 1
    sents = t_mod[i]
    #print(sents)
    for word in sents:
        if word in covid.keys():
            covid_prob *= covid[word]/sum(covid.values())
        else:
            covid_prob *= 0.001
        if word in sports.keys():
            sports_prob *= sports[word]/sum(sports.values()) 
        else:
            sports_prob *= 0.001
        if word in business.keys():
            business_prob *= business[word]/sum(business.values()) 
        else:
            business_prob *= 0.001
        if word in science.keys():
            science_prob *= science[word]/sum(science.values())
        else:
            science_prob *= 0.001
    class_condition.append([covid_prob,sports_prob,business_prob,science_prob])
    posterier = [covid_prob*covid_pr,sports_prob*sports_pr,business_prob*business_pr,science_prob*science_pr]
    index = posterier.index(max(posterier))
    expected_label.append(labels[index])
    
print(expected_label)
```

```python
check = 0
for i in range(test_size):
    if expected_label[i] == test_sample[i,0]:
        check += 1
```

```python
accuracy = (check/test_size)*100
print('Accuracy of Prediction : {}%'.format(accuracy))
```

```python
new_set = pd.read_csv('40.csv')
new_set = new_set.iloc[:,0].values
```

```python
new_sents = [ line.translate(punc).lower() for line in new_set ]
new_sents = [ [word for word in line.split() if word not in stop_words] for line in new_sents ]
```

```python
#I am trying to create vocabulary for each individual word in order to predict it correctly

voc = {}
for line in new_sents:
    for word in line:
        voc.update({word : {}})
        
for line in new_sents:
    for word in voc.keys():
        if word in line:
            for instance in line:
                if instance in voc[word].keys():
                    voc[word][instance] +=1
                else:
                    voc[word].update({instance : 1})
                    
```

```python
#Prior probablities
pr_prob = {word :0 for word in voc.keys()}
summ = 0

for word in voc.keys():
    count = 0
    for ins in voc[word].keys():
        count += voc[word][ins]
    summ = summ + count
    pr_prob[word] = count
    
pr_prob = [ val/summ for val in pr_prob.values() ]
pr_prob[:5] # prior probabities for each labels
```

```python
#implementation in test set
test = pd.read_csv('10.csv').iloc[:,0].values
test[:2]
```

```python
new_test = [ line.translate(punc).lower() for line in test ]
new_test = [ [word for word in line.split() if word not in stop_words] for line in new_test ]
new_test[:2]
```

```python
#class conditional probablities

labels = [ ins for ins in voc.keys() ]
voc_label = [ vals for vals in voc.values() ]
class_cond = []

for line in new_test:
    c_sent = [1 for s in voc.keys()]
    temp = []
    for word in line:
        prob = [0.001 for s in voc.keys()]
        for i in range(len(voc_label)):
            if word in voc_label[i].keys():
                prob[i] = voc_label[i][word]/sum(voc_label[i].values())
        temp.append(prob)
        #print(temp)
    for i in range(len(voc.keys())):
        for j in range(len(temp)):
            c_sent[i] *= temp[j][i]
    class_cond.append(c_sent)

#print(class_cond)

posterier = []

for ent in class_cond:
    post = [ ent[i]*pr_prob[i] for i in range(len(labels)) ]
    posterier.append(post)
    
#print(posterier)

last_word = []
confidence = []

for post in posterier:
    last_word.append(labels[post.index(max(post))])
    confidence.append(max(post))

i=0
for line in test:
    print("{} {} | confidence : {} ".format(line[:-5],last_word[i],confidence[i]))
    i = i+1
```
