```python
import pandas as pd
import numpy as np
import random
```

```python
dataset = pd.read_csv('Advertising Dataset.csv')
from sklearn.utils import shuffle
dataset = shuffle(dataset)
dataset
```

```python
training_set = dataset.iloc[:int(dataset.shape[0]*0.75)]
test_set = dataset.iloc[int(dataset.shape[0]*0.75):]
cols = len(training_set.iloc[0])
test_set
```

```python
y_train = training_set.loc[:,training_set.columns[cols-1]]
y_train = pd.DataFrame(y_train)
y_train
```

```python
x_train = training_set.loc[:,training_set.columns[1:cols-1]]
x_train
```

```python
x_mean = np.mean(x_train)
x_mean = pd.DataFrame(x_mean).T

y_mean = np.mean(y_train)

n=len(x_train)
```

```python
x_bar = x_train.copy()
y_bar = y_train.copy()

for i in range(n):
    x_bar.iloc[i] = x_train.iloc[i] - x_mean.iloc[0]
    y_bar.iloc[i] = y_train.iloc[i] - y_mean.iloc[0]
```

```python
denominator = np.linalg.inv((x_bar.T).dot(x_bar))
denominator
```

```python
numerator = (x_bar.T).dot(y_bar)
numerator = np.array(numerator)
numerator
```

```python
W = denominator.dot(numerator)
b = y_mean - np.array(x_mean.dot(W))[0]
print("Co-efficients of the hypothesis using formula are")
for i in range(len(W)):
    print(W[i][0])
print("Intercept of the hypothesis using formula : {}".format(b[0]))
W
```

```python
#using python library
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(x_train, y_train)
c = reg.coef_
print("Co-efficients of the hypothesis using formula are")
for i in range(len(c[0])):
    print(c[0][i])
print("Intercept of the hypothesis using library : {}".format(reg.intercept_[0]))
```

```python
x_test = test_set.loc[:,test_set.columns[1:cols-1]]
y_test = test_set.loc[:,test_set.columns[cols-1]]
y_test = pd.DataFrame(y_test)

def mean_error(x,y,w,b):
    m = len(x)
    error = np.array(y) - np.array(x).dot(w) - b[0]
    sqlerror = error**2
    mean_sql_error = sum(sqlerror)/m
    return mean_sql_error
    
mean_sql_error = mean_error(x_train,y_train,W ,b)
    
print("Mean square error is {} ".format(mean_sql_error[0]))
```

```python
# Ridge Regression

lam = [0.5,1,5,10,100,10000,50000,100000,500000,1000000] #lambda value was increase to get a curve while

w_ridge = []
b_ridge = []
mean_sql_errors = []

d = len(x_train.iloc[0])
for i in range(len(lam)):
    deno = np.linalg.inv((x_bar.T).dot(x_bar) + lam[i]*np.identity(d))
    numer = np.array((x_bar.T).dot(y_bar))
    w_ridge.append(deno.dot(numer))
    b_ridge.append(y_mean - np.array(x_mean.dot(w_ridge[i]))[0])
    mean_sql_errors.append(list(mean_error(x_test,y_test,w_ridge[i],b_ridge[i]))[0])

mean_coeff = [ np.mean(arr) for arr in w_ridge]
max_coeff = [np.max(arr) for arr in w_ridge]
print(mean_sql_errors)
print(mean_coeff)
print(max_coeff)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(lam,mean_coeff,'bo-')
plt.xlabel('Lambda')
plt.ylabel('Coefficients')
plt.title('Mean Coefficients Vs Lambda')
plt.show()
```

```python
plt.plot(lam,mean_sql_errors,'co-')
plt.xlabel('Lambda')
plt.ylabel('Mean Square Errors')
plt.title('Mean Square Errors Vs Lambda')
plt.show()
```

```python
plt.plot(lam,max_coeff,'go-')
plt.xlabel('Lambda')
plt.ylabel('Co-efficients')
plt.title('Max Co-efficients Vs Lambda')
plt.show()
```

```python
from sklearn.linear_model import Lasso
```

```python
coeff_ind = []
for val in lam[:5]: #sliced because we had to cosider only the first five values of lambda to find the most imp feature
    reg = Lasso(alpha = val)
    reg.fit(x_train, y_train)
    value = [bit for bit in reg.coef_]
    index  = value.index(max(value))
    coeff_ind.append(index)
    
most_imp = max(coeff_ind, key=coeff_ind.count)
most_imp
```

```python
imp_col = dataset.columns[most_imp+1] # +1 is done because data set also includes serial number but we are not including it
imp_col
```

```python
print("The most important feature is {} which is at {}th position according to Lasso Regression ".format(imp_col,most_imp))
```
