```python
from scipy.io import loadmat
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Lasso
```

```python
#importing the training set
train = loadmat('2010rainfall.mat')
train.keys()
```

```python
days=[]
for i in range(len(train['XR1'][0])):
    days.append(i+1)
len(days)
```

```python
empty=[]
location=[]
for i in range(len(train['XR1'])):
    location.append(i+1)
    empty.append(0)
len(location)    
```

```python
df=pd.DataFrame(data=train['XR1'],index=location,columns=days)
df.insert(0,'-1',empty,True)
df.insert(0,'-2',empty,True)
df
```

```python
test = loadmat('2011rainfall.mat')
test.keys()
```

```python
test_days=[]
for i in range(len(test['XR'][0])):
    test_days.append(i+1)

test_location=[]
for i in range(len(test['XR'])):
    test_location.append(i+1)
test_df=pd.DataFrame(data=test['XR'],index=test_location,columns=test_days)
test_df.insert(0,'-1',empty,True)
test_df.insert(0,'-2',empty,True)
test_df
```

```python
req_location = { 42 : "Mumbai", 158 : "Delhi", 299 : "Kharagpur" }

for s in req_location:
    y_train=np.array(df.iloc[s-1,2:].T)
    x_train=df.drop([s-1]).iloc[:,2:].values
    prev1 = df.iloc[s,1:123].values
    prev2 = df.iloc[s,0:122].values
    
    x_train = np.vstack((x_train,prev1))
    x_train = np.vstack((x_train,prev2))
    
    reg=LinearRegression() 
    reg.fit(x_train.T,y_train)
    
    
    y_test = np.array(test_df.iloc[s-1,2:].T)
    x_test = test_df.drop([s-1]).iloc[:,2:].values
    prev1 = test_df.iloc[s,1:123].values
    prev2 = test_df.iloc[s,0:122].values
    
    x_test = np.vstack((x_test,prev1))
    x_test = np.vstack((x_test,prev2))
    
    prediction = reg.predict(x_test.T)

    print('Mean squared error for {} is {} '.format(req_location[s],mean_squared_error(y_test, prediction , squared=True)))
```

# Q5-Q6

```python
#Importing the  train and test dataset
mat1 = loadmat('2010rainfall.mat')
df_X_train  = pd.DataFrame(mat1['XR1'])
df_y_train = pd.DataFrame(mat1['ZR1'])

mat2 = loadmat('2011rainfall.mat')
df_X_test = pd.DataFrame(mat2['XR'])
df_y_test = pd.DataFrame(mat2['ZR'])

X_train = df_X_train.iloc[:,:].values
X_train = X_train.T
y_train = df_y_train.iloc[:].values
y_train = y_train.T

X_test = df_X_test.iloc[:,:].values
X_test = X_test.T
y_test = df_y_test.iloc[:].values
y_test = y_test.T

```

```python
#Applying decision tree algorithm on the training set
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy')
classifier.fit(X_train, y_train)
```

```python
#Applying the decision tree algorithm on the test set
y_pred = classifier.predict(X_test)
```

```python
#Computing the accuracy of predictions
count = 0
for i in range(len(y_test)):
    if y_pred[i] == y_test[i]:
        count += 1
print("Percentage accuracy: {}".format(count*100/len(y_test)))
 
```

```python
#Retrieving the most important features
importances = classifier.feature_importances_
importances = [bit for bit in importances]
importances_max = importances
importances_max = sorted(importances_max, reverse = True)
importances_max = [bit for bit in importances_max if bit]
indices = [importances.index(element) for element in importances_max]
print(indices)
```
